### Build normalized histograms with kde plots to explore the distributions further. 
### Use only the continuous variables in the dataset to plot these visualizations. 

```python
for column in ['Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Weekly_Sales']:
    walmart[column].plot.hist(density = True )
    walmart[column].plot.kde(label = column )
    plt.legend()
    plt.show()
```

```python
# State your observations here
# Again, none of the histograms depict a normal distribution. Size, CPI and Fuel Price all have multiple modes. 
# Normalization will be required for these features and the high scale will need to be addressed in the weekly sales.
```

### Build joint plots to check for the linearity assumption between predictors and target variable

Let's use a slightly more advanced plotting technique in seaborn that uses scatter plots, distributions, kde and simple regression line - all in a single go. Its called a `jointplot`. [Here is the official doc. for this method](https://seaborn.pydata.org/generated/seaborn.jointplot.html). 

Here is how you would use it:

> **`sns.jointplot(x= <column>, y= <column>, data=<dataset>, kind='reg')`**

A joint plot will allow us to visually inspect linearity as well as normality assumptions as a single step. 

```python
for column in ['Store', 'Dept', 'Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']:
    sns.jointplot(x = column, y = "Weekly_Sales", data = walmart, kind='reg', label=column, joint_kws={'line_kws':{'color':'green'}})
#     sns.regplot(walmart[column], walmart.Weekly_Sales, label=column)
    plt.legend()
    plt.show()
```

```python
# Based on these graphs, we can discern that Store, Dept, Furl, Temp and Unemployment do not have a clear 
# relationship with weekly sales. The vertical lines of data points indicate that there is no linear relationship
# with weekly sales.
# Looking at the green lines, CPI and Size have very weak relationships with weekly sales. 
# CPI - negative, Size - positive
# Normality does not hold for these predictors.
```
